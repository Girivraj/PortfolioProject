Line graph: trying to track changes of a shorter or longer period of time. for smaller changes always prefer line graphs.

Bar graph: bar graph or horizontal bar graph have the same usecases.
Bar graph are used to compare changes between different groups or to track changes over time, however when trying to measure changes over time bar graphs are best when the changes are large.

Pie charts are used to compare parts of a whole, They do not show changes over time.
